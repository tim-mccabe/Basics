```python
# This file contains a few basic functions cleaning data and parsing dates to refer back to 
# when writing new code

# modules we'll use
import pandas as pd
import numpy as np
import seaborn as sns
import datetime

# read in the data
earthquakes = pd.read_csv("data/database.csv")

# set seed for reproducibility
np.random.seed(0)

earthquakes.head()
```

```python
# check the data type of our date column

earthquakes["Date"].dtype
```

```python
# some dates are not in the same format as the others

earthquakes[3378:3383]
```

```python
# see where else this occuance is in the data

date_lengths = earthquakes.Date.str.len()
date_lengths.value_counts()
```

```python
indicies = np.where([date_lengths == 24])[1]
print("Incicies with corrupted data:", indicies)
earthquakes.loc[indicies]
```

```python
# Convert our date columns to datetime

earthquakes.loc[3378, "Date"] = "02/23/1975"
earthquakes.loc[7512, "Date"] = "04/28/1985"
earthquakes.loc[20650, "Date"] = "03/13/2011"
earthquakes["date_parsed"] = pd.to_datetime(earthquakes["Date"], format = "%m/%d/%Y")
```

```python
# select the day of the month

day_of_the_month = earthquakes["date_parsed"].dt.day
```

```python
# plot the day of the month to check the date parsing

day_of_the_month = day_of_the_month.dropna()

sns.distplot(day_of_the_month, kde = False, bins = 31)
```
