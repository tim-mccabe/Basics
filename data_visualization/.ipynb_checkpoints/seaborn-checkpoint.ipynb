```python
# This file contains a few basic functions of seaborn creating differnt types of data visualizations to 
# refer back to when writing new code
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Read in and create new dataframes for each data set

# this data was provided by the courses offered on kaggels website for data visualization

spotify_df = pd.read_csv("data/spotify.csv", index_col = "Date", parse_dates = True)
museum_df = pd.read_csv("data/museum_visitors.csv", index_col = "Date", parse_dates = True)
ign_df = pd.read_csv("data/ign_scores.csv", index_col = "Platform")
candy_df = pd.read_csv("data/candy.csv", index_col = "id")
cancer_b_df = pd.read_csv("data/cancer_b.csv", index_col = "Id")
cancer_m_df = pd.read_csv("data/cancer_m.csv", index_col = "Id")
iris_df = pd.read_csv("data/iris.csv", index_col = "Id")
```

```python
# Creating a line chart of all data. the index will be the index of the data in this case

sns.lineplot(data = museum_df)
```

```python
# Creating a line chart of specific data from dataframe. the index will be the index of the data in this case

sns.lineplot(data=museum_df["Avila Adobe"], label="Avila Adobe")
```

```python
# setting the size of the chart that will be printed below the cell
plt.figure(figsize = (25, 6))

# creating a barchart 
sns.barplot(x = ign_df.index, y = ign_df["Racing"])
```

```python
# creating a heatmap. annot = True refers to having numbers show up on each square of heatmap

plt.figure(figsize = (10, 10))
sns.heatmap(data = ign_df, annot = True)
```

```python
# creating a scatter plot between two varables

sns.scatterplot(x = candy_df.index, y = candy_df["winpercent"])
```

```python
# creating a scatter plot between two varables with regression line

sns.regplot(x = candy_df.index, y = candy_df["winpercent"])
```

```python
# Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'
sns.scatterplot(x = candy_df.index, y = candy_df["winpercent"], hue = candy_df["chocolate"])
```

```python
# Color-coded scatter plot w/ regression lines
sns.lmplot(x = "pricepercent", y = "winpercent", hue = "chocolate", data = candy_df)
```

```python
# Scatter plot showing the relationship between 'chocolate' and 'winpercent'
sns.swarmplot(x = candy_df["chocolate"], y = candy_df["winpercent"])
```

```python
# Histograms for benign and maligant tumors

sns.distplot(a = cancer_b_df["Area (mean)"], label = "B", kde = False)
sns.distplot(a = cancer_m_df["Area (mean)"], label = "M", kde = False)
```

```python
# KDE plots for benign and malignant tumors

sns.kdeplot(data = cancer_b_df["Radius (worst)"], label = "B", shade = True)
sns.kdeplot(data = cancer_m_df["Radius (worst)"], label = "M", shade = True)
```

```python
# Change the style of the figure
sns.set_style("ticks")

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_df)
```

```python
# KDE plot 
sns.kdeplot(data=iris_df['Petal Length (cm)'], shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=iris_df['Petal Length (cm)'], y=iris_df['Sepal Width (cm)'], kind="kde")
```
