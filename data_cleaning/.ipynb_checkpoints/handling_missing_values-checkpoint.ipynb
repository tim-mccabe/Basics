```python
# This file contains a few basic functions cleaning data and handling missing data to refer back to when writing new code

# import moduels / libraries
import pandas as pd
import numpy as np

# read in data
sf_permits = pd.read_csv("data/Building_Permits.csv")

# set seed for reproducibility
np.random.seed(0)
```

```python
# print the first 5 row to see where data is missing

sf_permits.head()
```

```python
# find out how many are missing and what percent of the whole data set has missing values

missing_value_counts = sf_permits.isnull().sum()
total_cells = np.product(sf_permits.shape)
total_missing = missing_value_counts.sum()

percent_missing = (total_missing / total_cells) * 100
print(total_missing)
print(percent_missing)
```

```python
# drop all missing rows and see what the data would look like

sf_permits_with_na_rows_dropped = sf_permits.dropna()
sf_permits_with_na_rows_dropped
```

```python
# drop all missing columns and see what the data would look like

sf_permits_with_na_columns_dropped = sf_permits.dropna(axis = 1)
sf_permits_with_na_columns_dropped
```

```python
# fill missing values with the one that comes directly after it and then replace any remaning NaN's with 0

sf_permits_with_na_imputed = sf_permits.fillna(method = "bfill", axis = 0).fillna(0)
sf_permits_with_na_imputed
```
